```python
# Import necessary functions from IPython to play the audio file within the notebook.

from IPython.display import Audio, display

# Display and autoplay the downloaded audio file to ensure it was downloaded correctly.

display(Audio('audio.mp3', autoplay=True))
```

```python
# Import the pipeline function from the transformers library.

from transformers import pipeline
```

```python
# Initialize the whisper model for automatic speech recognition (ASR) using a medium-sized model.
# The 'device' parameter is set to 0 to utilize the first GPU. If no GPU is available, it defaults to CPU.

whisper = pipeline('automatic-speech-recognition', model = 'openai/whisper-medium', device = 0)
```

```python
# Perform speech recognition on the downloaded audio file and store the transcribed text.

text = whisper('audio.mp3')
```
